# 1. 당뇨병 데이터를 가지고 머신러닝 5가지 분류를 수행.

# 필요한 라이브러리 불러오기

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

# 데이터 불러오기

```python
# 데이터 불러오기
url = 'https://raw.githubusercontent.com/MyungKyuYi/AI-class/main/diabetes.csv'
df = pd.read_csv(url)

# 데이터 확인
df
```

# 결측치 확인 및 제거


```python
# 결측치 확인
df.isnull().sum()
```

# 테스트, 훈련 데이터로 분류

```python
# 테스트, 훈련 데이터 나누기 (X = 당뇨병인지 확인할 수 있는 여러 요소들, Y = 당뇨병 유무 결과)
X = df.drop('Outcome', axis=1)
y = df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

# SVM 분류

```python
# SVM (Support Vector Machine) 분류
clf_svm = SVC(random_state=0)
clf_svm.fit(X_train, y_train)
pred_svm = clf_svm.predict(X_test)
print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

# LR 분류

```python
# LR (Logistic Regression) 분류
clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

# RF 분류

```python
# RF(Random Forest) 분류
clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

# DT 분류


```python
# DT(Decision Tree) 분류
clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

# KNN 분류

```python
clf_knn = KNeighborsClassifier()
clf_knn.fit(X_train, y_train)
pred_knn = clf_knn.predict(X_test)

print ("\n--- K-Nearest Neighbors Classifier ---")
print (accuracy_score(y_test, pred_knn))
print (confusion_matrix(y_test, pred_knn))
```

```python
# 데이터 불러오기
df = pd.read_csv(url)
df
```

```python
# 테스트, 훈련 데이터 나누기 (X = bmi를 추정할 수 있는 요소들, Y = bmi)
X = df.drop(['Outcome', 'BMI'], axis=1) # outcome 삭제, bmi는 예측요소이므로 제거
y = df['BMI'] # 예측하고자 하는 대상은 bmi
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
# 신경망 모델 구축
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
```

```python
# 모델 컴파일
model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mse'])

# 모델 학습
model.fit(x=X_train, y=y_train, epochs=100, batch_size=32,validation_data= (X_test,y_test))

# 테스트 데이터로 예측 수행
y_pred = model.predict(X_test)

# 예측 결과와 실제 값 출력
mse = mean_squared_error(y_test, y_pred)
print(f'Neural Network MSE: {mse:.2f}')  # 평균 제곱 오차 출력
```
